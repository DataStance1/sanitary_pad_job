#### How I handled the multiple responses in this case.

Dealing with variables with multiple responses in spss can be tedious. So I decided to get around the usual method of turning each value to a variable and then filling up with a Yes or a No. This was going to be very tedious because the data had five columns with these multiple responses and each of the columns have about nine unique values. So, this implies creating nine more variables for each column and filling them up with Yes or No for the 1000 observations. This will result to 45 extra columns. The data will be so wide and this task will be extremely tedious. 


- So I decided to try another method.

I used the Unnest(string_to_array) in sql to separate the variables that had the comma separated values. What this function does is that it adds extra observations for each cell that has more than one unique values separated by a comma. 

To do this, I had to load this data to PgAdmin. I encountered three problems why trying to do that

1. My pgAdmin binary path was at the default setting. I had to change it to C:\Program Files\PostgreSQL\15\bin.

2. I had this error message- ERROR: extra data after last expected column. This was because the data, which is a CSV file with comma as the delimiter, I was trying to load into pgAdmin had columns with comma separated values. So pgAdmin most likely treated the both commas as indicating when to start a new columm. To resolve this I had to re-save the data as a CSV file with semi colon as the delimiter.

3. I had this error message- duplicate key value violates unique constraint "sanitary_pad_pkey". The identification number column I used as the primary key in the table, had a duplicate value, hence pgAdmin could not process it . I had to drop the column and used pandas df.reset_index() method to create a column column that I used as the primary key.

Below is the query that does the job. This query selects every column in the table, It unnests the columns q3b, q3c, q3d, q4a, q4b by adding a new row(observation) to the table for each cell with more than one unique value separated by a comma

```python
%%sql
SELECT question_id, age, religion, q1a, q1b, q1c, q1d, q2a, q2b, q2c, q2d, q3a, 
        UNNEST(STRING_TO_ARRAY(q3b, ',')) as q3b,
        UNNEST(STRING_TO_ARRAY(q3c, ',')) as q3c,
        UNNEST(STRING_TO_ARRAY(q3d, ',')) as q3d,
        q4,
        UNNEST(STRING_TO_ARRAY(q4a, ',')) as q4a,
        UNNEST(STRING_TO_ARRAY(q4b, ',')) as q4b
FROM sp.sanitary_pad
```

-- To Check if the output of the query is consistent with the original data
--I unnested the all the columns and then used a cte to count the table and group be q4a. 
--Then I unnested only q4a and used a cte to count the table and group by q4a. I then compared the two outputs.
--The two outputs are the same.
--Just that the unnesting all the columns added some null values to the q4a column. Which is inconsequential.


```python
%%sql
with cte as (
SELECT question_id, age, religion, q1a, q1b, q1c, q1d, q2a, q2b, q2c, q2d, q3a, 
        UNNEST(STRING_TO_ARRAY(q3b, ',')) as q3b,
        UNNEST(STRING_TO_ARRAY(q3c, ',')) as q3c,
        UNNEST(STRING_TO_ARRAY(q3d, ',')) as q3d,
        q4,
        UNNEST(STRING_TO_ARRAY(q4a, ',')) as q4a,
        UNNEST(STRING_TO_ARRAY(q4b, ',')) as q4b
FROM sp.sanitary_pad
    )
    SELECT COUNT(*), q4a
    FROM cte
    GROUP BY q4a
```
